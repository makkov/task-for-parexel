```python
#Импортируем данные
import pandas
data = pandas.read_csv('irisdata.txt', header=None)
```

```python
#Назначим столбцам имена
data.columns=["sepal length", "sepal width", "petal length", "petal width", "class"]
```

```python
data.head()
```

```python
#Функция получения данных по значению класса ириса
def getDataSetByTypeOfClass(dataName, columnName, typeOfClass):
    return dataName.loc[dataName[columnName] == typeOfClass]

#Функция подсчета объектов в данных 
def getCountNumber(dataName):
    return ["N = " + str(len(dataName))]

#Функция получения статистической информации для набора данных по определенному столбцу
def getStatisticalInformation(dataName, columnName):
    info = []
    info.append(dataName[columnName].count())
    info.append(round(dataName[columnName].mean(), 1))
    info.append(round(dataName[columnName].min(), 1))
    info.append(round(dataName[columnName].median(), 1))
    info.append(round(dataName[columnName].max(), 1))
    info.append(round(dataName[columnName].std(), 1))
    return info

#Функция подсчета количества объектов с определенными параметрами длины
def getLenghtInfo(dataName, columnName):
    totalNumber = dataName[columnName].count()
    info = [
    str((dataName[columnName] < 5).sum()) + 
        " (" + str(round(float((dataName[columnName] < 5).sum())/totalNumber, 3) * 100) + ")",
    str(((dataName[columnName] >= 5) & (dataName[columnName] < 6)).sum()) +
        " (" + str(round(float((((dataName[columnName] >= 5) & (dataName[columnName] < 6)).sum()))/totalNumber, 3) * 100) + ")",
    str(((dataName[columnName] >= 6) & (dataName[columnName] < 7)).sum()) +
        " (" + str(round(float(((dataName[columnName] >= 6) & (dataName[columnName] < 7)).sum())/totalNumber, 3) * 100) + ")",
    str((dataName[columnName] >= 7).sum()) +
        " (" + str(round(float((dataName[columnName] >= 7).sum())/totalNumber, 3) * 100) + ")"
    ]
    return info

#Функция подсчета количества объектов с определенными параметрами ширины
def getWidthInfo(dataName, columnName):
    totalNumber = dataName[columnName].count()
    info = [
    str((dataName[columnName] < 3).sum()) + 
        " (" + str(round(float((dataName[columnName] < 3).sum())/totalNumber, 3) * 100) + ")",
    str(((dataName[columnName] >= 3) & (dataName[columnName] < 3.5)).sum()) +
        " (" + str(round(float(((dataName[columnName] >= 3) & (dataName[columnName] < 3.5)).sum())/totalNumber, 3) * 100) + ")",
    str(((dataName[columnName] >= 3.5) & (dataName[columnName] < 4)).sum()) +
        " (" + str(round(float(((dataName[columnName] >= 3.5) & (dataName[columnName] < 4)).sum())/totalNumber, 3) * 100) + ")"
    ]
    return info

#Функция представления информации по определенному набору данных
def getGeneralInfo(dataTypeOfIris):
    countNumber = getCountNumber(dataTypeOfIris)
    statisticalInformationOnSepalLength = getStatisticalInformation(dataTypeOfIris, 'sepal length')
    sepalLengthInfo = getLenghtInfo(dataTypeOfIris, 'sepal length')
    statisticalInformationOnSepalWidth = getStatisticalInformation(dataTypeOfIris, 'sepal width')
    sepalWidthInfo = getWidthInfo(dataTypeOfIris, 'sepal width')
    return countNumber + [''] + statisticalInformationOnSepalLength + [''] + sepalLengthInfo + [''] + statisticalInformationOnSepalWidth + [''] + sepalWidthInfo
```

```python
#Создадим набор данных для каждого вида ирисов и получим по ним информацию в виде массива
dataTotal = data
infoTotal = getGeneralInfo(dataTotal)
dataSetosa = getDataSetByTypeOfClass(data, 'class', 'Iris-setosa')
infoSetosa = getGeneralInfo(dataSetosa)
dataVersicolor = getDataSetByTypeOfClass(data, 'class', 'Iris-versicolor')
infoVersicolor = getGeneralInfo(dataVersicolor)
dataVirginica = getDataSetByTypeOfClass(data, 'class', 'Iris-virginica')
infoVirginica = getGeneralInfo(dataVirginica)
```

```python
#Создадим таблицу на основе полученных массивов
matrix = {'Iris-Setosa': infoSetosa,
         'Iris-Versicolor': infoVersicolor,
         'Iris-Virginica': infoVirginica,
         'Total': infoTotal}
df = pandas.DataFrame(data = matrix)
```

```python
#Переименуем строки таблицы
df = df.rename(index={0: '',
                      1: 'SEPAL LENGTH [CM]',
                      2: 'N',
                      3: 'MEAN',
                      4: 'MIN',
                      5: 'MEDIAN',
                      6: 'MAX',
                      7: 'STANDARD DEVIATION',
                      8: 'SEPAL LENGTH (%)',
                      9: '< 5',
                      10: '>= 5 AND < 6',
                      11: '>= 6 AND < 7',
                      12: '>= 7',
                      13: 'SEPAL WIDTH [CM]',
                      14: 'N',
                      15: 'MEAN',
                      16: 'MIN',
                      17: 'MEDIAN',
                      18: 'MAX',
                      19: 'STANDARD DEVIATION',
                      20: 'SEPAL WIDTH (%)',
                      21: '< 3',
                      22: '>= 3 AND < 3.5',
                      23: '>= 3.5 AND < 4'})
```
